```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import LogLocator
from matplotlib.ticker import FixedLocator, FixedFormatter
date = "0617"
month = "06"
time = "11"
# Step 1: Load the shapefile (ensure you download a shapefile for California counties)
shapefile_path = "data\CA_Counties\CA_Counties.shp"  # Replace with your shapefile path
gdf = gpd.read_file(shapefile_path)
input_file = f"Result/2019{month}_county/total_carbon_{date}_27_county_{time}.csv"
# Step 2: Load the emission data
emission_data = pd.read_csv(input_file)  # Replace with your CSV file path

# Step 3: Merge the emission data with the GeoDataFrame
# Ensure the county names in your CSV match the shapefile's county name column
# Adjust 'NAME' below to match your shapefile's county name column
gdf = gdf.merge(emission_data, left_on="NAMELSAD", right_on="county")

# Step 4: Plot the map with color gradients
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
gf = gdf.plot(
    column="carb_emi per capita (metric tonnes per million)",  # Column to base color gradient on
    cmap="RdYlGn_r",             # Color map (e.g., 'OrRd' for orange-red gradient)
    legend=True,             # Add legend
    legend_kwds={"label": "Carbon Emission per capita(metric tonnes/ million people)"},
    edgecolor="black",       # Add black borders to counties
    ax=ax,
    norm=matplotlib.colors.LogNorm(vmin=min(emission_data["carb_emi per capita (metric tonnes per million)"]), vmax=1400))

for cax in fig.axes:
    if cax != ax:  # The map is ax; the colorbar is the other one
        color_bar = cax
        break
tick_values = [1, 5, 10, 50, 100, 500, 1000]
#color_bar = fig.axes[1]
color_bar.yaxis.set_major_locator(FixedLocator(tick_values))
color_bar.yaxis.set_major_formatter(FixedFormatter([str(val) for val in tick_values]))
color_bar.tick_params(labelsize=14)
#lg = ax.get_legend()
#lg.set_fontsize(14)
ax.set_title("Carbon emission per capita On 06/17/2019 at 11am", fontsize=16)
ax.axis("off")  # Remove axis

# Save the figure
fig_name = f"Result/2019{month}_county/carbon_emi_per_capita_{time}.png"
#plt.legend(fontsize=14)
plt.savefig(fig_name, dpi=300, bbox_inches="tight")

plt.show()
```
